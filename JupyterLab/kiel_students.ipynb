# Number of Students in Kiel

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("kiel_students.csv", sep=";")
print(df.head())
```

```python
fig, ax = plt.subplots(figsize=(10, 5))

x = df["Wintersemester"]
y = df["Christian-Albrechts-Universitaet"]

plt.title("Students in Kiel (CAU)", size="x-large")
plt.ylabel("Number of Students", size="x-large")
plt.xlabel("Winter Semester", size="x-large")

plt.plot(y, "*-", markersize=6, linewidth=1, color='b', label="Unfälle")

ax.set_xticks(range(len(x)))
ax.set_xticklabels(x, rotation='vertical')

plt.show()
```
